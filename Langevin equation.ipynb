```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-0.2, 2, 100)

#zero inertia approx
b = 20
o = 8
x0 = 1
v0 = 1
y = x0*np.exp((-o**2*x)/b)

#full overdamped langevin equation
obar = ((b**2)/4 - o**2)**(1/2)
y_1 = np.exp(-b*x/2)*(x0*np.cosh(obar*x) + (1/obar)*(v0 + b*x0/2)*np.sinh(obar*x))
```

```python
fig = plt.figure(figsize = (10, 5))
plt.ylabel('x(t)')
plt.xlabel('time (s)')
plt.ylim(-0.1,1.5)
plt.grid('true')
plt.plot(x, y , color='red',label='Zero inertia approximation')
plt.plot(x, y_1, color='blue', label='Full overdamped Langevin equation')
plt.legend()
plt.savefig('graphtraj.png')
plt.show()
```
